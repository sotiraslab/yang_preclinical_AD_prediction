```python
wdir = "/home/b.y.yang/sotiraslab/BradenADLongitudinalPrediction"
features_df = pd.read_csv(os.path.join(wdir, "code/exploratory/a4_treatment/pet_analysis/features_A4_treatment.csv"))  # features from all subjects
ohe_sex = OneHotEncoder(categories = [["F", "M"]], drop="if_binary", sparse_output = False)
features_df["sex_M"] = ohe_sex.fit_transform(features_df["sex"].values.reshape(-1, 1))
a4_df = features_df[features_df["site"] == "A4"]
```

```python
model_dir = "/ceph/chpc/shared/aristeidis_sotiras_group/b.y.yang_scratch/BradenADLongitudinalPrediction/3_TrainTestModel/simplified_experiments/base"
```

```python
a4_df_input = a4_df.copy()
for m, t in it.product(["svm"], [1,2,3,4,5]):
    print(f"{m}, {t}")
    
    col = f"y_score.{m}.{t}"
    model_dict = joblib.load(os.path.join(model_dir, f"avn__{m}__{t}yr__site.joblib"))
    model = model_dict["trained_models"]["A4"]["model"].best_estimator_

    if m == "rf":
        y_score = model.predict_proba(a4_df_input)[:,1]
    else:
        y_score = model.decision_function(a4_df_input)
    a4_df.loc[:,col] = y_score
```
